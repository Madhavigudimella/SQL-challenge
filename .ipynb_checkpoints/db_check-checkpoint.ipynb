```python
# Dependencies

import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import pandas as pd
import psycopg2
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
from pword import password
location = f'postgresql://postgres:{password}@localhost:5432/homework'
engine = create_engine(location)
connection = engine.connect()
```

```python
#Test to ensure data us being retrived properly
inspector = inspect(engine)
inspector.get_table_names()
```

```python
#Read salaries data into dataframe
session = Session(engine)

salaries_df = pd.read_sql_table('salaries',engine)
salaries_df.describe()
```

```python
#create a histogram to visualize the most common salary ranges for employees.
plt.hist(salaries_df['salary'], bins=10, align='right', color='blue', edgecolor='black')
plt.xlabel('# of employees')
plt.ylabel('Salary of employees')
plt.title('Salary Distribution')
```

```python
#Create a bar chart of average salary by title.

#Read employees data into dataframe
employees_df = pd.read_sql_table('employees',engine)

#Read titles data into dataframe
titles_df = pd.read_sql_table('titles',engine)

#Merge employee and salary data frames
emp_salary = employees_df.merge(salaries_df, how="left", on='emp_no')

#Merge title data frames
emp_sal_title = emp_salary.merge(titles_df, how="left", left_on='emp_title_id', right_on='title_id')

#Group by Title
emp_salary_group = emp_sal_title.groupby(['title']).aggregate({'salary':'mean'})
emp_salary_group

# Create a bar chart based off of the group series from before
salary_chart = emp_salary_group.plot(kind='bar')

# Set the xlabel and ylabel using class methods
salary_chart.set_xlabel("Employee Title")
salary_chart.set_ylabel("Average Salary")
salary_chart.set_title("Average Employee Salary by Title")


plt.show()
plt.tight_layout()
```

Above histogram indicates huge variance between salaries and the bar graph indicates that managers are being paid less than staff. Both these conclusions indicate that the data is questionable.
